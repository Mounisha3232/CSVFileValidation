# Filling Null values

```python
for i in range(len(l)):
    if(l[i]):
        data[d[i]].fillna(data[d[i]].mode()[0],inplace = True) 
```

```python
dl=[]
for j in range(len(l)):
    dl.append((type(data[d[j]][1])))
```

# LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in range(len(dl)):
    if dl[i]==str:
        data[d[i]]= le.fit_transform(data[d[i]])
```

# Finding dublicate rows

```python
dataarray=data.to_numpy()
```

```python
dataarray
```

```python
dataarray[0]
```

```python
droplist=[]
for k in range(len(dataarray)):
    for j in range(k+1,len(dataarray)):
        if(list(dataarray[k])==list(dataarray[j])):
            droplist.append(j)           
```
