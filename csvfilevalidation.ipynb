# identifying the datatype

# LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in range(len(dl)):
    if dl[i]==str:
        data[coloum[i]]= le.fit_transform(data[coloum[i]])
```
